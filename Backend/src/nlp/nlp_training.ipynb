```python
def evaluate_model(nlp, eval_data):
    y_true = []
    y_pred = []

    for sample in eval_data:
        text = sample["text"]
        true_ings = set(i.upper() for i in sample["true_ingredients"])

        doc = nlp(text)
        pred_ings = set(ent.text.upper() for ent in doc.ents if ent.label_ == "INGREDIENT")

        all_ings = true_ings.union(pred_ings)

        for ing in all_ings:
            y_true.append(1 if ing in true_ings else 0)
            y_pred.append(1 if ing in pred_ings else 0)

    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    return precision, recall, f1
```

```python

nlp_before = spacy.blank("en")

precision_b, recall_b, f1_b = evaluate_model(nlp_before, EVAL_DATA)

print("📉 BEFORE TRAINING")
print(f"Precision: {precision_b:.2f}")
print(f"Recall:    {recall_b:.2f}")
print(f"F1 Score:  {f1_b:.2f}")
```

```python
MODEL_PATH = "ingredient_model"

if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError("Train the model first using train_ner.py")

nlp_after = spacy.load(MODEL_PATH)
```

```python
precision_a, recall_a, f1_a = evaluate_model(nlp_after, EVAL_DATA)

print("📈 AFTER TRAINING")
print(f"Precision: {precision_a:.2f}")
print(f"Recall:    {recall_a:.2f}")
print(f"F1 Score:  {f1_a:.2f}")
```
